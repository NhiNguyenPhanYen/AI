```python
import copy
import math

X = "X"
O = "O"
EMPTY = None

HUMAN = O
AI = X


def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]


def player(board):
    count = sum(cell is not None for row in board for cell in row)
    return X if count % 2 == 0 else O


def actions(board):
    return {(i, j)
            for i in range(3)
            for j in range(3)
            if board[i][j] == EMPTY}


def result(board, action):
    (i, j) = action
    if board[i][j] is not EMPTY:
        raise Exception("Ô này đã được đánh!")

    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board


def winner(board):
    lines = []

    lines.extend(board)
    lines.extend([[board[i][j] for i in range(3)] for j in range(3)])
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2 - i] for i in range(3)])

    for line in lines:
        if line[0] is not None and line.count(line[0]) == 3:
            return line[0]
    return None


def terminal(board):
    if winner(board) is not None:
        return True
    return all(cell is not EMPTY for row in board for cell in row)


def utility(board):
    w = winner(board)
    if w == X:
        return 1
    elif w == O:
        return -1
    return 0


def minimax(board):
    if terminal(board):
        return None

    turn = player(board)

    if turn == AI:
        value = -math.inf
        best_move = None
        alpha = -math.inf
        beta = math.inf

        for action in actions(board):
            v = min_value(result(board, action), alpha, beta)
            if v > value:
                value = v
                best_move = action
            alpha = max(alpha, value)

        return best_move

    else:
        value = math.inf
        best_move = None
        alpha = -math.inf
        beta = math.inf

        for action in actions(board):
            v = max_value(result(board, action), alpha, beta)
            if v < value:
                value = v
                best_move = action
            beta = min(beta, value)

        return best_move


def max_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = -math.inf
    for action in actions(board):
        v = max(v, min_value(result(board, action), alpha, beta))
        if v >= beta:
            return v
        alpha = max(alpha, v)
    return v


def min_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action), alpha, beta))
        if v <= alpha:
            return v
        beta = min(beta, v)
    return v



def print_board(board):
    print("\nBàn cờ:")
    for row in board:
        print([" " if cell is None else cell for cell in row])
    print()




board = initial_state()

print("Bạn là O, AI là X")
print("Nhập tọa độ từ 0 đến 2 (ví dụ: 1 2)\n")

while not terminal(board):
    print_board(board)

    if player(board) == HUMAN:
        while True:
            try:
                i = int(input("Nhập dòng (0-2): "))
                j = int(input("Nhập cột (0-2): "))
                if (i, j) in actions(board):
                    move = (i, j)
                    break
                else:
                    print("Ô không hợp lệ, nhập lại!")
            except:
                print("Nhập sai, thử lại!")

    else:
        move = minimax(board)
        print("AI chọn:", move)

    board = result(board, move)


print_board(board)

w = winner(board)
if w == HUMAN:
    print(" Bạn thắng!")
elif w == AI:
    print(" AI thắng!")
else:
    print(" Hòa!")
```
